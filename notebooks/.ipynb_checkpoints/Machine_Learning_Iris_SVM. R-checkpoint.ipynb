## Density Plots by Class (Base R)

````r
# Prepare input and output
x <- dataset[, 1:4]
y <- dataset[, 5]

# Colors for classes
colors <- c("red", "green", "blue")
names(colors) <- levels(y)

# Set up 2x2 plotting layout
par(mfrow = c(2, 2))

for (i in 1:4) {
  # Compute y-axis limit (max density value across all classes for attribute i)
  y_max <- max(sapply(levels(y), function(cl) max(density(x[y == cl, i])$y)))
  
  # Compute x-axis range for attribute i
  x_range <- range(x[, i])
  
  # Plot the first class density to set up the plot
  plot(density(x[y == levels(y)[1], i]),
       main = paste("Density of", names(x)[i], "by Species"),
       xlab = names(x)[i], ylab = "Density",
       col = colors[levels(y)[1]], lwd = 2,
       xlim = x_range,
       ylim = c(0, y_max)
  )
  
  # Add density lines for other classes
  for (cl in levels(y)[-1]) {
    lines(density(x[y == cl, i]), col = colors[cl], lwd = 2)
  }
  
  # Add legend
  legend("topright", legend = levels(y), col = colors, lwd = 2)
}

# Reset plotting layout
par(mfrow = c(1, 1))
````

##  Evaluating Algorithms with 10-Fold Cross-Validation

We perform 10-fold cross-validation to estimate model accuracy on unseen data.  
Three algorithms are evaluated: Linear Discriminant Analysis (LDA), Classification and Regression Trees (CART), and k-Nearest Neighbors (kNN).

# Evaluating a Simple Logistic Regression Model on the Iris Dataset

Since my working environment has limited packages and does not include popular machine learning libraries like `caret`, `MASS`, or `e1071`, I am implementing a basic classification model using base R functions only.


## Making Predictions and Evaluating Accuracy

- The model outputs probabilities of each sample belonging to the "setosa" class.
- A threshold of 0.5 is used to assign the predicted class.
- Finally, I calculate the accuracy as the proportion of correct predictions on the test set.


````r
# Load iris dataset
data(iris)
dataset <- iris

# Convert Species to a binary factor (e.g., setosa vs others)
dataset$SpeciesBinary <- factor(ifelse(dataset$Species == "setosa", "setosa", "other"))

# Split into train and test sets (80/20 split)
set.seed(42)
train_index <- sample(seq_len(nrow(dataset)), size = 0.8 * nrow(dataset))
train_data <- dataset[train_index, ]
test_data <- dataset[-train_index, ]

# Train logistic regression model (base R glm)
model <- glm(SpeciesBinary ~ Sepal.Length + Sepal.Width + Petal.Length + Petal.Width,
             data = train_data, family = binomial)

# Predict on test data (probabilities)
probs <- predict(model, test_data, type = "response")

# Convert probabilities to classes with threshold 0.5
predicted <- ifelse(probs > 0.5, "setosa", "other")

# Calculate accuracy
actual <- test_data$SpeciesBinary
accuracy <- mean(predicted == actual)
cat("Logistic Regression Accuracy:", round(accuracy * 100, 2), "%\n")
# --- Manual k-Nearest Neighbors (k=3) ---
euclidean_dist <- function(x1, x2) sqrt(sum((x1 - x2)^2))

knn_predict <- function(train_x, train_y, test_x, k = 3) {
  apply(test_x, 1, function(test_point) {
    distances <- apply(train_x, 1, function(train_point) euclidean_dist(train_point, test_point))
    nearest_indices <- order(distances)[1:k]
    nearest_labels <- train_y[nearest_indices]
    names(sort(table(nearest_labels), decreasing = TRUE))[1]
  })
}

train_x <- train_data[, 1:4]
train_y <- train_data$SpeciesBinary
test_x <- test_data[, 1:4]

knn_preds <- factor(knn_predict(train_x, train_y, test_x, k = 3), levels=levels(dataset$SpeciesBinary))
knn_accuracy <- mean(knn_preds == test_data$SpeciesBinary)
cat("kNN (k=3) Accuracy:", round(knn_accuracy * 100, 2), "%\n")

# --- Nearest Centroid Classifier ---
centroids <- aggregate(. ~ SpeciesBinary, data = train_data[, c(1:4, 6)], mean)

predict_centroid <- function(sample) {
  distances <- apply(centroids[, -1], 1, function(centroid) sqrt(sum((sample - centroid)^2)))
  centroids$SpeciesBinary[which.min(distances)]
}

centroid_preds <- apply(test_data[, 1:4], 1, predict_centroid)
centroid_preds <- factor(centroid_preds, levels=levels(dataset$SpeciesBinary))
centroid_accuracy <- mean(centroid_preds == test_data$SpeciesBinary)
cat("Nearest Centroid Accuracy:", round(centroid_accuracy * 100, 2), "%\n")
# Accuracy values
accuracies <- c(
  Logistic_Regression = log_accuracy,
  kNN = knn_accuracy,
  Nearest_Centroid = centroid_accuracy
)

# Convert to percentages for better readability
accuracies_percent <- round(accuracies * 100, 2)

# Create a barplot
barplot(
  accuracies_percent,
  main = "Model Accuracy Comparison",
  ylab = "Accuracy (%)",
  col = c("skyblue", "lightgreen", "salmon"),
  ylim = c(0, 110),
  beside = TRUE
)

# Add text labels above bars to show exact values
text(
  x = seq_along(accuracies_percent),
  y = accuracies_percent,
  label = paste0(accuracies_percent, "%"),
  pos = 3, # above the bar
  cex = 0.8,
  col = "black"
)

````

# Comparison of Machine Learning Algorithms on Iris Dataset in R
After training multiple models, the Linear Discriminant Analysis (LDA) model emerged as the best performer based on cross-validation accuracy.

````r
# Load dataset
data(iris)
dataset <- iris

# Split data into training and test sets
set.seed(7)
train_idx <- sample(1:nrow(dataset), size = 0.8 * nrow(dataset))
train_set <- dataset[train_idx, ]
test_set <- dataset[-train_idx, ]

# Euclidean distance function
euclidean_distance <- function(x1, x2) {
  sqrt(sum((x1 - x2)^2))
}

# Custom kNN function
custom_knn <- function(train_data, train_labels, test_data, k = 3) {
  predictions <- sapply(1:nrow(test_data), function(i) {
    distances <- apply(train_data, 1, function(x) euclidean_distance(x, test_data[i, ]))
    nearest_idx <- order(distances)[1:k]
    nearest_labels <- train_labels[nearest_idx]
    # Majority vote
    names(sort(table(nearest_labels), decreasing = TRUE))[1]
  })
  return(predictions)
}

# Prepare numeric data and labels
train_features <- train_set[, 1:4]
train_labels <- train_set$Species
test_features <- test_set[, 1:4]

# Predict on test set
knn_predictions <- custom_knn(train_features, train_labels, test_features, k = 3)

# Confusion matrix
conf_matrix <- table(Predicted = knn_predictions, Actual = test_set$Species)
print(conf_matrix)

# Accuracy calculation
accuracy <- sum(diag(conf_matrix)) / sum(conf_matrix)
cat(sprintf("Custom kNN Accuracy: %.2f%%\n", accuracy * 100))
````
